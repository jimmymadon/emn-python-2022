```python
import pandas as pd

dataframe = pd.read_csv('dirtydata.csv')
print(dataframe)
```

```python
# Cleaning NaN / None / Null values
# dropna() creates a NEW dataframe and cleans all rows with none / NaN values.
non_null_dataframe = dataframe.dropna()
print(non_null_dataframe)
```

```python
# If you want to overwrite the original dataframe, use inplace parameter.
dataframe.dropna(inplace=True)
print(dataframe)
```

```python
dataframe2 = pd.read_csv('dirtydata.csv')
dataframe2.fillna(130, inplace=True)
print(dataframe2)
```

```python
dataframe3 = pd.read_csv('dirtydata.csv')
# Remember: A COLUMN in a dataframe is a Pandas SERIES
print(type(dataframe3['Calories']))
# You can run mean(), median() and mode() functions on any series
calories_mean = dataframe['Calories'].mean()
# fillna() can be run on a Series too like below
dataframe3['Calories'].fillna(calories_mean, inplace=True)
print(dataframe3)
```

```python
# Cleaning Wrong Format data
dataframe4 = pd.read_csv('dirtydata.csv')
# The below line creates a new column/Series in our dataframe4
dataframe4['New Dates'] = pd.to_datetime(dataframe4['Date'])
print(dataframe4)

```

```python
dataframe4.dropna(subset=['New Dates'], inplace=True)
print(dataframe4)
```

```python
# Replacing WRONG data values
dataframe5 = pd.read_csv('dirtydata.csv')
print(dataframe5)
```

```python
mode = dataframe5['Duration'].mode()
print(mode)
dataframe5.loc[7, 'Duration'] = mode[0]
print(dataframe5)
```

```python
dataframe6 = pd.read_csv('dirtydata.csv')
print(dataframe6.index)
print(type(dataframe6.index))
# Loop through every row in a dataframe
mode = dataframe6['Duration'].mode()[0]
for row in dataframe6.index:
    if (dataframe6.loc[row, 'Duration']) > 120:
        # Replace the incorrect value with the mode of the column
        dataframe6.loc[row, 'Duration'] = mode

print(dataframe6)
        
```

```python
dataframe7 = pd.read_csv('dirtydata.csv')
print(dataframe7.index)
print(type(dataframe7.index))
# Loop through every row in a dataframe
for row in dataframe7.index:
    if (dataframe7.loc[row, 'Duration']) > 120:
        # Replace the incorrect value with the mode of the column
        dataframe7.drop(row, inplace=True)

print(dataframe7)
```

```python
# Remove DUPLICATE rows
dataframe8 = pd.read_csv('dirtydata.csv')
print(dataframe8.duplicated(subset=['Duration']))
print(dataframe8.duplicated())
```

```python
dataframe8.drop_duplicates(inplace=True)
print(dataframe8)
```
