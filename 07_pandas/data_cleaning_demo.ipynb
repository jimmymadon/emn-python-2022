```python
import pandas as pd

dataframe = pd.read_csv('dirtydata.csv')
print(dataframe)
```

```python
# Cleaning NaN / None / Null values
# dropna() creates a NEW dataframe and cleans all rows with none / NaN values.
non_null_dataframe = dataframe.dropna()
print(non_null_dataframe)
```

```python
# If you want to overwrite the original dataframe, use inplace parameter.
dataframe.dropna(inplace=True)
print(dataframe)
```

```python
dataframe2 = pd.read_csv('dirtydata.csv')
dataframe2.fillna(130, inplace=True)
print(dataframe2)
```

```python
dataframe3 = pd.read_csv('dirtydata.csv')
# Remember: A COLUMN in a dataframe is a Pandas SERIES
print(type(dataframe3['Calories']))
# You can run mean(), median() and mode() functions on any series
calories_mean = dataframe['Calories'].mean()
# fillna() can be run on a Series too like below
dataframe3['Calories'].fillna(calories_mean, inplace=True)
print(dataframe3)
```

```python
# Cleaning Wrong Format data
dataframe4 = pd.read_csv('dirtydata.csv')
# The below line creates a new column/Series in our dataframe4
dataframe4['New Dates'] = pd.to_datetime(dataframe4['Date'])
print(dataframe4)

```

```python
dataframe4.dropna(subset=['New Dates'], inplace=True)
print(dataframe4)
```
